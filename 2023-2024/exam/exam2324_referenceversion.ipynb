🥖 Baguettes' price usually is 1.15€. However, inflation 📈 is now 8.5%.

Write a function that, considering the inflation, returns the total price for a number of baguettes given as argument.
Then, create a program that uses this function to display the cost of 32 baguettes.

Beware: this function must be easily adaptable to different inflation and baguette price values.

<br/>

🇫🇷 Les baguettes 🥖 coûtent normalement 1,15€. L’inflation 📈 a désormais atteint 8,5 %.
Écrire une fonction qui retourne le prix total réel incluant l’inflation pour un nombre de
baguettes donné en paramètre.

Ensuite, créez un programme qui utilise cette fonction pour afficher le résultat du prix de
32 baguettes.

Attention : cette fonction doit être facilement adaptable à d'autres valeurs d'inflation et d'autres prix de baguette

````python
def price(numberOfBaguettes, baguettePrice=1.15, inflation=8.5):
	"""
	function that calculates how much you will have to pay for a number of baguettes
	"""
	flatPrice = numberOfBaguettes * baguettePrice
	inflationImpact = flatPrice * inflation / 100
	finalPrice = flatPrice + inflationImpact
	return finalPrice

print(price(32))


# ===== output from chatGPT
def calculate_total_cost(num_baguettes, baguette_price, inflation_rate):
	total_cost = num_baguettes * (1 + inflation_rate) * baguette_price
	return total_cost

# Example usage with 32 baguettes
num_baguettes = 32
baguette_price = 1.15
inflation_rate = 0.085  # 8.5%

total_cost = calculate_total_cost(num_baguettes, baguette_price, inflation_rate)

print(f"The total cost for {num_baguettes} baguettes is {total_cost:.2f}€.")

````

#### programming(3).png **TODO!**  🌶 🌶 🌶

Create a class named `Library` which contains the list of books.

This `Library` possesses the following methods:
- a method to find books by their year of publication
- a method to get the list of publishers without duplicates
- a method to add a book from a title, a year, a publisher and an author

By the way, this book is a MUST-READ!! 🤓

🇫🇷 Créez une classe nommée `Bibliotheque` qui contient la liste de livres.

Cette classe `Bibliotheque` possède les méthodes suivantes :
- une méthode pour trouver les livres par leur année de publication
- une méthode pour obtenir la liste des éditeurs sans redondance
- une méthode pour ajouter un livre à partir d'un titre, d'une année, d'un éditeur et d'un auteur

D'ailleurs, ce livre est un incontournable ! Vous devriez VRAIMENT le lire 🤓

````python
class Library():
	"""
	class to represent a library
	"""
	def __init__(self, books):
		self.books = books

	def findByYear(self, year):
		"""find books from the library by their year of publication"""
		byYear = []
		for book in self.books:
			if book["Year-Of-Publication"] == year:
				byYear.append(book)
		return byYear

	def getPublishers(self):
		"""retrieves the list of different publishers from the library"""
		publishers = []
		for book in self.books:
			publishers.append(book["Publisher"])
		return list(set(publishers))

	def addBook(self, title, author, year, publisher):
		""" add a book to the library """
		self.books.append({
			"Book-Title": title,
			"Book-Author": author,
			"Year-Of-Publication": year,
			"Publisher": publisher
		})

````
