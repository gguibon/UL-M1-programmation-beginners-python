````python
def getCharactersWithWands(wand=True):
  """
  get characters with or without a wand depending on the parameter
  By default, returns the ones with a wand.
  """
  res = []
  if wand:
    for c in characters:
      try:
        if c["Wand"].lower() != "unknown":
          res.append(c)
      except:
        pass
  else:
    for c in characters:
      if c["Wand"] is None:
        res.append(c)
      elif c["Wand"].lower() == "unknown":
        res.append(c)
  return res
````

Create a class to represent a `Spell` with the following attributes:

It should try to retrieve the random number from a range from the `manadist` attribute and print the effect, the mana consumed, and the light from the spell. Here is a sample output:

ðŸ‡«ðŸ‡· CrÃ©ez une classe pour reprÃ©senter un `Sort` (ou `Spell` en anglais) avec les attributs suivants :
- nom (*name*)
- incantation (*incantation*)
- effect (*effect*)
- light (*light*): oÃ¹ toutes les valeurs nulles (`None`) doivent devenir des chaÃ®nes de caractÃ¨res (String) `"None"`
- types (*types*)
- manadist: attribut qui contient le dictionnaire suivant:

Ensuite ajoutez la mÃ©thode suivante :
- `launch` pour dÃ©finir le processus de lancement du sort. Cette mÃ©thode doit rÃ©cupÃ©rer un nombre alÃ©atoire Ã  partir d'une fenÃªtre dÃ©finie dans `manadist` et afficher son effet, le nombre de mana consumÃ©e et la lumiÃ¨re du sort. Si la couleur n'est pas prÃ©sente dans `manadist`, alors un nombre au hasard entre 0 et 100 doit Ãªtre sÃ©lectionnÃ©. 
Voici un exemple de sortie :
```
Prevents people above or below a certain age from access to a target with 3 mana consumed (Blue)
```

ðŸ’¡ Astuce :
- utilisez `import random` et `random.randint` pour rÃ©cupÃ©rer un entier alÃ©atoire

````python
import random
class Spell():
  """
  class to represent a spell
  """
  def __init__(self, spell_dict):
    self.name = spell_dict["Name"]
    self.incantation = spell_dict["Incantation"]
    self.effect = spell_dict["Effect"] 
    try:
      self.light = spell_dict["Light"]
    except:
      self.light = "None"
    self.types = spell_dict["Types"]
    self.manadist = {"blue": [0, 10], "yellow": [5, 15], "purple": [30, 35] }

  def launch(self):
    """
    launches the spell and indicates the number of mana consumed.
    Returns the number of mana consumed.
    """
    manarange = self.manadist.get(self.light.lower(), [0,100])
    mana_consumed = random.randint(manarange[0], manarange[1])
    print(self.effect, f"with {mana_consumed} mana consumed ({self.light})")
    return mana_consumed
````

````python
spells_instances = []
for s in spells:
  spells_instances.append(Spell(s))
````
