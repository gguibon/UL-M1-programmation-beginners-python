# <font color="teal">**Recursive Functions**</font>

A recursive function is a function that called itself in it.

It can serve multiple purposes, and can be quite difficult to grasp. In this section you will see multiple usages for recursive functions. There are many usages so this is not an exhaustive representation of recursivity. However, it is important to understand it. If you do understand recursive functions, then you should not have any issue understanding all the previous notions we tackled in this course. If so, this only means one thing: you did not pratice enough 🙀

Let's dive into it by creating a function to explore a list of firstnames and put them into uppercase.

````python
super_students = ["camille", "enzo", "ilan", "pin-xun", "tian"]

def names_to_upper(names, index=0):
	"""
	This function is a recursive function to modify a list of string. It makes each string into an uppercase and goes by one element at a time.
	Arguments:
		names (list): list of string names
		index (int): current index to process. Default value: 0
	"""
	# we modify the element of the list of names at the given index. By using .upper() we put all the characters into uppercase
	names[index] = names[index].upper()
	# now thaht we are done, we can increment the index and put it as the next index to process
	next_index = index + 1
	# we need to verify if the next index exists in the list. A good way is to just check that this index does not exceed the length of the list of names
	if next_index < len(names):
		# if the index exists we first print a message from the function to us (we are the creators of this cute function ;) )
		print(f"Hi my dear creator. I 🤖 just processed the index {index} and will not call myself with the index {next_index}.")
		print(f"The current list is {names}")
		# now we can call the same function, with different arguments: the list of names is update at the current index, and we specify the next index to process
		names_to_upper(names, index=next_index)
	else:
		# if the next index does not exist in the list we print a farewell message from the function. Sayonara cute function, you served us well 😥
		print("Dear creator, I am your function 'names_to_upper' 🤖 and I just finished processing the list you asked me to process. Look at it. Are you proud? 🥹🥹")
		# we directly modify the list of super_students, so that we can use it outside of this function
		super_students = names

# now that our recursive function is defined, we can call it with just one line
names_to_upper(super_students)

# let's print the updated list by our recursive function
print(super_students)
````

Now, you would protest: "But we could do this really easily with a for loop!". Indeed, the previous recursive function could be simplified as:

````python
super_students = ["camille", "enzo", "ilan", "pin-xun", "tian"]

def upper_students(names):
	"""
	This function put all the strings into a list of string as uppercase
	Arguments:
		names (list): a list of strings
	"""
	for i in range(len(names)):
		super_students[i] = names[i].upper()

# let's call our function
upper_students(super_students)
# then print the modified list
print(super_students)
````

# <font color="teal">**Basic Role Playing Game**</font>

* Objective: construct a textual role playing game. The game is only text with choices, conditions to verify the choices and player status.

## <font color="teal">Rules</font>
* Player advance from a room to another by textually selecting one of the rooms.
* Player starts with 200 hp (health points)
* A bad decision costs the player to lose 25 hp.
* For each actions, display it from a list of Strings
* Represent data as dictionaries or a big nested dict()
* Use functions to remove code duplicates and reduce the line numbers
* Add status and health modification function
* **Separate the code from the scenario** (external JSON)
* **Write past choices into an external file**

````python
# Example with dict  and  ba sic functions and health management
# Placeholders for you to insert requested code
player = dict()
print("Hello Player One, what's your name?")
player['name'] = input()
player['health'] = 200
print("Welcome", player['name'])


def retrieveEventsFromJson(jsonPath):
    """
    This function reads an external json file containing the events infos and returns a dictionary.
    """
    eventsFile = open(jsonPath, 'r') # open the file in 'r' read mode
    eventsDict = json.load(eventsFile) # load the file content into a dict
    eventsFile.close() # do not forget to close the files
    return eventsDict

def writePastChoices( past ):
    """ function that write the past choices to a file .txt """
    pastFile = open('pastChoices.txt', 'a') # we append the choices
    pastFile.write('\n=======\n======\n\n') # append some markers to differenciates the runs
    for p in past: # we iterate over the choices to append them
        pastFile.write(p + '\n') # '\n' means new line
    pastFile.close()

def removeHealth(amount):
    ''' Remove some health from the player. And check if he died. '''
    player['health'] -= amount
    if player['health'] <= 0:
        return False
    else:
        return True

pastChoices = []
events = retrieveEventsFromJson('events.json') # we use the function to get events dict from the json file


def askAction(event, choices):
    '''
    Ask an action from the user. Stay alive while the action is not recognized.
    This is a recursive function 😛
    '''
    choice = ''
    while choice not in choices:
        print('player status', player)
        choice = input()
        try: # using exception instead of list checking
            print(event[choice]['message']) # if this does not work -> exception
            if 'hploss' in event[choice]:
                alive = removeHealth(event[choice]['hploss'])
                if not alive : break
            pastChoices.append(choice)
            if('paths' in event[choice]):
                askAction(event[choice]['paths'], list(event[choice]['paths'].keys()) )
        except:
            print('WRONG : Possible choices', choices )
    return True

print('''Your are in a tiny room. Humidity fills the air but your stomach reminds you that you are very hungry.
        You are in front of two doors. Behind the first one you can hear muffled voices.
        Behind the second one you can smell something intriguing.''')
print('Which door do you choose? Type', list(events.keys())[0], 'for the first room and', list(events.keys())[1],
              'for the second room.')

askAction(events, list(events.keys()) )

print('Here are your remaining stats', player)
writePastChoices( pastChoices ) # we use the function to write the pastChoices into a file.
````
