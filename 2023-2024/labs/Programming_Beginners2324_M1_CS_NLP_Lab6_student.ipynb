# <font color="teal">**Objects**</font>

A `class` is a way to define new objects. When you create a class you already define rules and properties for future objects. If you look at the image above, you will see that the biscuits are objects, they are instances of the cake mold. Thus, it is really convenient as you will be able to define a class once, and then create objects from this class as many times as you need to.


Let's take another example. Lets' consider us! Humans! We are not humans. We are.... instances of the `class` human.

Let's create the human class.

````python
class Human():
  """
  This is a class to represent Humans.
  """
  def __init__(self, firstname, lastname, sex, age=0):
    """
    The init method refers to the initialization of the object. Thus, it will automatically be called.
    Here we can do some starter things. The most common ones are the properties definition (here: firstname, sex, height, etc.)
    Age is set to zero by default because a baby is not born already old ;)
    """
    # we define the property firstname and initialize it with the firstname argument
    self.firstname = firstname
    # we define the property lastname and initialize it with the lastname argument
    self.lastname = lastname
    # we define the property age and initialize it with the age argument (which is 0 by default)
    self.age = age
    # we define the property sex and initialize it with the sex argument
    self.sex = sex
    # we define the property stomach and initialize it with an empty list
    self.stomach = []
    # now that we defined the properties, we will be able to use them in methods containing the 'self' argument
    # (a method is a function inside a class such as the following)

  def eat(self, food):
    """
    This method have to be called from an initialized object. Hence it needs the self argument.
    Usually all humans know how to eat. So we create a method for this.
    """
    # here we append the given food into the stomach. We access the stomach variable using the self. approach because it is a property
    # note that we could consider giving a limit to the number of food a human can eat. But eating is fun so let's not limit it :)
    self.stomach.append(food)
````

Now that we created a basic class to represent humans, let's create some humans!!

````python
# let's represent a famous movie maker ("2046", "Chungking Express", etc.)
# we initialize the Human() class so it automatically runs the __init__ method
wong_karwai = Human("Kar-Wai", "Wong", "Male", age=64)
# let's represent a famous musician ("Tajabone" may be his best song)
ismael_lo = Human("Ismaël", "Lô", "Male", age=66)
````

Traditionally, class properties should have a `set` (to change the property) and a `get` (to retrieve the property) method. As you can see, Python does not necessarily needs this as `wong_karwai.firstname` is already doing what the `get` method would do. However, it is still considered a good pratice for setting a property by using an instanciated object as it will allow you to control the way the property is defined.

To put it simply, let's look at this example which redefines the `Human` class.

````python
class Human():
  """
  This is a class to represent Humans.
  """
  def __init__(self, firstname, lastname, sex, age=0):
    """
    The init method refers to the initialization of the object. Thus, it will automatically be called.
    Here we can do some starter things. The most common ones are the properties definition (here: firstname, sex, height, etc.)
    Age is set to zero by default because a baby is not born already old ;)
    """
    self.firstname = firstname
    self.lastname = lastname
    self.age = age
    self.sex = sex
    self.stomach = []

  def set_firstname(self, text):
    """
    This method allows us to define how to set the firstname outside of the init method.
    This will automatically put the string value `text` in lowercase with the first letter being capitalized.
    Thus "gaËl", will become "Gaël"
    Convenient isn't it?
    """
    # first we modify the incoming text
    text = text[0].upper() + text[1:].lower()
    # then we put it as the firstname value
    self.firstname = text

  def get_firstname(self):
    """
    This method allows to retrieve the property firstname.
    """
    # right now this method is a good practice but sometimes you will want to control the way values are retrieved.
    return self.firstname

  def eat(self, food):
    """
    This method have to be called from an initialized object. Hence it needs the self argument.
    Usually all humans know how to eat. So we create a method for this.
    """
    self.stomach.append(food)
````

Now that our class is better defined to control access (get) and modification (set) to the `firstname` property, let's use it 😛

````python
# we define again Wong Kar-Wai but with the update Human class
wong_karwai = Human("Kar-Wai", "Wong", "Male", age=64)

# now we use print the firstname using the getter
print(wong_karwai.get_firstname())
````

But wait! It did not change anything?!

Indeed we only retrieved the firstname and did not use the setter to control the firstname modification. Let's change the firstname using our setter method.

````python
# we use the setter method to modify the firstname
wong_karwai.set_firstname("scOobYdoO")

# then we print it
print(wong_karwai.get_firstname())
````

See? The firstname is automatically defined using the correct case. Awesome isn't it? 🤓

# <font color="teal">**Class as a Toolbox**</font>

One usage of the classes, is to define **static** methods. Static methods are often utility methods. These are methods which cannot access the instance property because they are not linked to an instance. This means you can access these methods directly without having to instance an object. the total opposite of what we did with the `Human` class.

Let's check some static class usage by defining a toolbox 🧰. A basic example of such toolbox is a calculator: we don't need properties, only a set of convenient methods.

````python
class Calculator():
  """
  Here is a "toolbox" usage of the OOP. Only a set of static methods. Which means we do not need an init method.
  """

  def do_sum(*args):
    """
    get the sum of numbers
    """
    return sum(args)

  def do_avg(list_of_num):
    """
    get the average of a list of numbers
    """
    return sum(list_of_num) / len(list_of_num)

  def abs_with_rest(num):
    """
    convenient method which yields a tuple with the absolute value and the decimals.
    there are multiple ways to do it, this one is a naive one
    """
    return ( abs(num), int(str(num).split(".")[1]) )
````

Now let's use this calculator.

````python
print( Calculator.do_sum(29, 4, 3) )
````

# <font color="teal">**Basic Role Playing Game**</font>

* Objective: construct a textual role playing game. The game is only text with choices, conditions to verify the choices and player status.

## <font color="teal">Rules</font>
* Player advance from a room to another by textually selecting one of the rooms.
* Player starts with 200 hp (health points)
* A bad decision costs the player to lose 25 hp.
* For each actions, display it from a list of Strings
* Represent data as dictionaries or a big nested dict()
* Use functions to remove code duplicates and reduce the line numbers
* Add status and health modification function
* **Separate the code from the scenario** (external JSON)
* **Write past choices into an external file**

````python
# first we need to download the json file for our example simple scenario
! wget https://gitlab.com/gguibon/idmc-storage/-/raw/main/events.json
````

````python
import json
# Example using object oriented programming

class myGame():
  def __init__(self, jsonPath):
      """
      initialization
      we start by reading the scenario from the json file
      then we initialize some player values

      please note that we could have also created a class for the player :)
      """
      with open(jsonPath, 'r') as f:
        self.events = json.load(f)
      self.pastChoices = []

      self.player = dict()
      print("Hello Player One, what's your name?")
      self.player['name'] = input()
      self.player['health'] = 200
      print("Welcome", self.player['name'])

  def writePastChoices(self):
      """
      function that write the past choices to a file .txt
      since we have properties, this function does not need an argument anymore
      """
      pastFile = open('pastChoices.txt', 'a') # we append the choices
      pastFile.write('\n=======\n======\n\n') # append some markers to differenciates the runs
      for p in self.pastChoices: # we iterate over the choices to append them
          pastFile.write(p + '\n') # '\n' means new line
      pastFile.close()

  def removeHealth(self, amount):
      ''' Remove some health from the player. And check if he died. '''
      self.player['health'] -= amount
      if self.player['health'] <= 0:
          return False
      else:
          return True

  def askAction(self, event, choices):
      '''
      Ask an action from the user. Stay alive while the action is not recognized.
      This is a recursive function 😛
      See how I added self. everytime I wanted to access a own method or property?
      '''
      choice = ''
      while choice not in choices:
          print('player status', self.player)
          choice = input()
          try: # using exception instead of list checking
              print(event[choice]['message']) # if this does not work -> exception
              if 'hploss' in event[choice]:
                  alive = self.removeHealth(event[choice]['hploss'])
                  if not alive : break
              self.pastChoices.append(choice)
              if('paths' in event[choice]):
                  self.askAction(event[choice]['paths'], list(event[choice]['paths'].keys()) )
          except:
              print('WRONG : Possible choices', choices )
      return True

  def start(self):
      print(self.events["message"])
      self.askAction(self.events['paths'], list(self.events["paths"]))
      # the game is over so we print the summary
      print("you followed the path:", self.pastChoices)
      print('Here are your remaining stats', self.player)
      self.writePastChoices() # we use the method to write the pastChoices into a file.

game = myGame("events.json")
game.start()
````
