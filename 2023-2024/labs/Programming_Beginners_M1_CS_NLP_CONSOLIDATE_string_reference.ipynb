```python
me = me.lower()
print(me)
```

Split the `me` variable based on the space. Use the `.split()` function to do so. Put it into a variable then print it.

```python
names = me.split(" ")
print(names)
```

Join the values of the variable based on a comma using the `"".join()` function and put it into a variable. then print it.

```python
me = ", ".join(names)
print(me)
```

Split the following citation to extract a list of words. Then print it.

```python
citation = "Your time is limited, so don't waste it living someone else's life. Don't be trapped by dogma – which is living with the results of other people's thinking."

print(citation.split(" "))
```

There is something wrong, commas and dots are not separated from the words. Hence, modify the citation to split the tokens (i.e. words, punctuation, etc.). 
To do so, combine the `split()` function and some `replace()` function. Then  print the ist of tokens.

```python
tokens = citation.replace(".", " .")
tokens = tokens.replace(",", " ,")
tokens = tokens.replace("'", " '")
tokens = tokens.split(" ")
print(tokens)
```

There is a issue here, some words are not in lowercase. Put all the tokens in lowercase. then print the modified list.

```python
for i in range(len(tokens)):
	tokens[i] = tokens[i].lower()

print(tokens)
```

Join the tokens into a single string where each token is separate by a space. Put it into a variable then print it.

```python
tokens = " ".join(tokens)
print(tokens)
```

Change the following String and put one of two character as uppercase. Such as "hello" becomes "HeLlO" for instance.

```python
# this is the longest French word
word = "anticonstitutionnellement"

res = ""
for i, c in enumerate(word):
	if (i % 2) == 0: res = res + c
	else: res = res + c.upper()

print(res)
```

Transform this string variable into a correct sentence.

```python
world = "thisIsSuchAWonderfulWorld"

words = []
idx = 0
w = ""
for c in world:
	if c.isupper() and idx+1 < len(world):
		words.append(w)
		w = ""
		idx+=1
	w = w+c.lower()

print( " ".join(words) )
```
